#A03. Desafio 01: Plotar as médias, medianas e modas de algumas notas, com uma linha reta vertical.

```
dados_sem_notas_nulas["NU_NOTA_MT"].mode()
```

```
dsnn = dados_sem_notas_nulas["NU_NOTA_MT"]
mean_dsnn = dsnn.mean()
median_dsnn = dsnn.median()
mode_dsnn = dsnn.mode()[0]
plt.figure(figsize = (12,8))
plt.axvline(mean_dsnn, color = "red")
plt.axvline(median_dsnn, color = "green")
plt.axvline(mode_dsnn, color = "blue")
plt.legend(['Mean', 'Mediam', 'Mode'])
sns.histplot(dados_sem_notas_nulas, x = "NU_NOTA_MT", 
             kde = True, color = ['grey'])
```

```
dados_sem_notas_nulas["NU_NOTA_LC"].mode()
```

```
dsnn = dados_sem_notas_nulas["NU_NOTA_LC"]
mean_dsnn = dsnn.mean()
median_dsnn = dsnn.median()
mode_dsnn = dsnn.mode()[0]
plt.figure(figsize = (16,8))
plt.axvline(mean_dsnn, color = "red")
plt.axvline(median_dsnn, color = "green")
plt.axvline(mode_dsnn, color = "blue")
plt.legend(['Mean', 'Mediam', 'Mode'])
sns.histplot(dados_sem_notas_nulas, x = "NU_NOTA_LC", 
             kde = True, color = ['grey'])
```
